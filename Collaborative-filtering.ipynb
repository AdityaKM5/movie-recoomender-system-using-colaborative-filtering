```python
pt=filtered.pivot_table(index='original_title',columns='userId',values='rating')
```

```python
pt.isnull().sum()
```

```python
pt.fillna(0,inplace=True)
```

```python
import ast
from sklearn.metrics.pairwise import cosine_similarity
```

```python
col_sim= cosine_similarity(pt)
```

```python
col_sim.shape
```

```python
np.where(filtered.index=='Toy Story')[0]
```

```python
def recommend(movie_name):
    ind= np.where(pt.index==movie_name)[0][0]
    distances = sorted(list(enumerate(col_sim[ind])),key= lambda x:x[1],reverse=True)[1:6]
    l=[]
    for i in distances:
        x=movie[movie['original_title']== pt.index[i[0]]]['tmdbId'].values[0]
        print(x)
        #l.append(x)
        l.append(pt.index[i[0]])
    return l
```
